```python
#class

class A:
    pass

a=A()
a.b='c'
print(a.b)
print(isinstance(a, A)) #a가 A의 인스턴스가 맞는지 확인
```

```python
#class 멤버 변수

class A:
    def __init__(self, a, b):
        self.a=a
        self.b=b

a1=A('에이', '비')
print(a1.a, a1.b)

a2=A('에이2', '비2')
print(a2.a, a2.b)
```

```python
#class 메소드

class BlackBox:
    def __init__(self, name, price):
        self.name=name
        self.price=price
    def set_travel_mode(self, min):
        print(min,'분 동안 여행모드 on')

b1=BlackBox('블랙박스', '100000원')
print(b1.name, b1.price)
b1.set_travel_mode(10)
```

```python
#class self

class BlackBox:
    def __init__(self, name, price):
        self.name=name
        self.price=price
    def set_travel_mode(self, min):
        print(f'{self.name} {min},분 동안 여행모드 on')

b1=BlackBox('b1', 20000)

b1.set_travel_mode(10) #똑같음
BlackBox.set_travel_mode(b1,10) #똑같음
```
