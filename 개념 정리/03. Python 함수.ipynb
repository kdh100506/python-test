```python
#함수 전달값
'''
def 함수명(전달값):
    수행할 문장
'''
#여러개가 사용이 가능(콤마로 구분), 함수 내에서만 사용

def A(a):
    print('A',a)

A(10)
```

```python
#예시

def show_price(hair_cut_price):
    print('커트 가격은',hair_cut_price,'원 입니다.')

show_price(15000)
```

```python
#함수 변환값
'''
def 함수명(전달값):
    수행할 문장
    return 값
'''
#여러개 변환 가능(콤마로 구분/튜플), 변환 되는 즉시 함수 탈출 

def a():
    return 10

b=a()
print(b)
```

```python
#예시

def hair_cut_price(vip):
    if vip==True:
        print('커트 가격은 15000원 입니다.')
    else:
        print('커트 가격은 20000원 입니다.')

hair_cut_price(True)
hair_cut_price(False)
```

```python
#함수 기본값
'''
def 함수명(전달값=기본값):
    수행할 문장
'''

def A(a=True):
    if a==True:
        print('true')
    else:
        print('false')

A(True)
A(False)
A()
```

```python
#예시

def hair_cut_price(vip=False):
    if vip==True:
        print('커트 가격은 15000원 입니다.')
    else:
        print('커트 가격은 20000원 입니다.')

hair_cut_price(True)
hair_cut_price(False)
hair_cut_price()
```

```python
#함수 키워드 값
'''
def 함수명(변환값/키워드):
    수행할 문장
'''

def A(a=False, b=False, c=False):
    print('a는',a,'이고, b는',b,'이고, c는',c,'이다.')

A()
A(True)
A(True, False, True)
A(c= True)
```

```python
#예시

def hair(hair_style='시스루', downpum='안', hair_cleaning='안'):
    print('오늘 머리는',hair_style,'로 할게요')
    print('다운펌은',downpum,'하시고, 머리감기는',hair_cleaning,'하실게요')

hair()
    
```

```python
#함수 가변인자
'''
def 함수명(*전달값 -> 가변인자):
    수행할 문장
'''
#가변인자는 전달값의 제일 마지막에 딱 한번만 쓸수있음

def A(*a):
    for b in a:
        print(b)

A(10,11,12)
```

```python
#예시

def visit(today, *customers):
    print(today)
    for customer in customers:
        print(customer)

visit('2026.03.10', 'A', 'B', 'C')
visit('2026.03.11', 'D', 'E')
```
