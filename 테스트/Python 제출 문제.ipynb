```python
#1번

student=1
print('세 과목의 점수에 전부 0을 입력하시면 종료됩나다.')
while True:
    sub1_point=int(input('제 1과목의 점수'))
    sub2_point=int(input('제 2과목의 점수'))
    sub3_point=int(input('제 3과목의 점수'))
    if sub1_point>0 or sub2_point>0 or sub3_point>0:
        point_total=(sub1_point+sub2_point+sub3_point)
        point_average=(sub1_point+sub2_point+sub3_point)/3
        print(student,'번째 학생 총점:',point_total,', 평균:',point_average)
        student+=1
    else:
        print('총',student-1,'명을 성적 처리하였습니다.')
        break
```
