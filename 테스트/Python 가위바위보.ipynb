```python
while True:
    a=input('1번째')
    b=input('2번째')
    if (a=='가위' and b=='바위') or (a=='바위' and b=='보') or (a=='보' and b=='가위'):
        print('2번째 사람이 승리 하였습니다.')
        break
    elif (a=='바위' and b=='가위') or (a=='보' and b=='바위') or (a=='가위' and b=='보'):
        print('1번째 사람이 승리 하였습니다')
        break
    else:
        print('무승부 입니다.')
```

```python
import random

while True:
    a = input('가위/바위/보 중 하나를 입력하세요: ')
    b = random.choice(['가위', '바위', '보'])
    
    print('컴퓨터:', b)

    if (a=='가위' and b=='바위') or (a=='바위' and b=='보') or (a=='보' and b=='가위'):
        print('컴퓨터가 승리했습니다.')
        break
    elif (a=='바위' and b=='가위') or (a=='보' and b=='바위') or (a=='가위' and b=='보'):
        print('플레이어가 승리했습니다.')
        break
    else:
        print('무승부 입니다.')
```
