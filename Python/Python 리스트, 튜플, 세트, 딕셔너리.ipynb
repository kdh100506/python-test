```python
# 리스트

my_list=[1, 2, 3, 'A', 'B', 'C', True, False]

print(my_list)
print('특징: 순서가 있음, 중복을 허용, 값을 수정할수 있음.')
```

```python
#튜플

my_tuple=(1, 2, 3)

print(my_tuple)
print('특징: 순서가 있음, 중복을 허용, 수정 불가능, 데이터 변경을 막을때 사용')
```

```python
#세트

my_set={1, 2, 3}

print(my_set)
print('순서가 없음, 중복을 허용하지 않음')
```

```python
#딕셔너리

my_dictionary={'A':1,'B':2}

print(my_dictionary)
print('key:value형식, 순서 없음, key는 중복이 안됨 value는 가능')
```
