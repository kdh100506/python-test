```python
# 계정 생성 & 로그인

print('계정 생성하기')
name = input("Name Set:")
password = input("Password Set:")
print('Name:',name)
print('Password:',password)

x=False
print('계정 로그인하기')
while x==False:
    name_input = input('Name:')
    password_input = input('Password:')
    print('Name:',name_input)
    print('Paasword:',password_input)
    if name_input == name and password_input == password:
        print('인증되었습니다')
        x=True
    else:
        print('Name 또는 Password가 틀렸습니다.')
        print('다시 로그인 하십시오.')
        x=False
```

```python
# 팔벌려뛰기 개수 세기 (for 반복문)

a=int(input('팔벌려뛰기 개수:'))
b=1

for i in range(a):
    print('팔벌려뛰기',b,'개')
    b+=1
```

```python
# 팔벌려뛰기 개수 세기 (while 반복문)

a=int(input('팔벌려뛰기 개수:'))
b=1

while b<=a:
    print('팔벌려뛰기',b,'개')
    b+=1
```

```python
# 리스트를 튜플로 변환

a=('사과','바나나','포도')
print(a)
b=list(a)
print(b)
b.append('딸기')
print(b)
a=tuple(b)
print(a)
```

```python
# 리스트를 세트로 변환후 리스트로 변환 (리스트의 중복값이 사라짐 / 순서 X)

a=['사과','바나나','포도','포도']
print(a)
b=set(a)
print(b)
a=list(b)
print(a)
```

```python
# 리스트를 딕셔너리로 변환후 리스트로 변환 (리스트의 중복값이 사라짐 / 순서 O)

a=['사과','바나나','포도','포도']
print(a)
b=dict.fromkeys(a)
print(b)
a=list(b)
print(a)
```
